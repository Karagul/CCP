```python
def eurex_colect(quarter):
    
    #quarter = format: 'Q3-2017'
    
    ccp = 'eurex'
    reverse = quarter.split('-')
    reverse = reverse[1] + '_' + reverse[0]
    
    #basic
    b_file = pd.read_excel('../CCP/raw_data/' + quarter.upper() + '/' + ccp + '-' + 
                           quarter.lower() + '.xlsx', sheet_name='AggregatedDataFile')
    #files in use
    a_file = pd.read_excel('../CCP/raw_data/' + quarter.upper() + '/' + ccp + '-' + 
                           quarter.lower() + '.xlsx', 
                           sheet_name='CCP_DataFile_4_4a')
    c_file = pd.read_excel('../CCP/raw_data/' + quarter.upper() + '/' + ccp + '-' + 
                           quarter.lower() + '.xlsx', 
                           sheet_name='CCP_DataFile_6_1')
    d_file = pd.read_excel('../CCP/raw_data/' + quarter.upper() + '/' + ccp + '-' + 
                           quarter.lower() + '.xlsx', 
                           sheet_name='CCP_DataFile_6_2')
    e_file = pd.read_excel('../CCP/raw_data/' + quarter.upper() + '/' + ccp + '-' + 
                           quarter.lower() + '.xlsx', 
                           sheet_name='CCP_DataFile_18_2')
    g_file = pd.read_excel('../CCP/raw_data/' + quarter.upper() + '/' + ccp + '-' + 
                           quarter.lower() + '.xlsx', 
                           sheet_name='CCP_DataFile_23')
    ex_rate = pd.read_excel('../CCP/raw_data/kursy-walutowe-CCP.xlsx', sheet_name='Arkusz1')
    
    #final file
    try:
        f_file = pd.DataFrame({'CCP':ccp.capitalize()}, index=[0])
        f_file['quarter'] = quarter
        f_file['clearning service'] = 'aggregated'
        f_file['currency'] = b_file['Currency']
        f_file['4.1.1'] = b_file['4_1_1']
        f_file['4.1.2'] = b_file['4_1_2']
        f_file['4.1.3'] = b_file['4_1_3']
        f_file['4.1.4'] = b_file['4_1_4']
        f_file['4.1.5'] = b_file['4_1_5']
        f_file['4.1.6'] = b_file['4_1_6']
        f_file['4.1.7'] = b_file['4_1_7']
        f_file['4.1.8'] = b_file['4_1_8']
        f_file['4.1.9'] = b_file['4_1_9']
        f_file['4.1.10'] = b_file['4_1_10']
    
    #for Q1-2016 & Q4-2015 -> they are different
    except:
        f_file = pd.DataFrame({'CCP':ccp.capitalize()}, index=[0])
        f_file['quarter'] = quarter
        f_file['clearning service'] = 'aggregated'
        f_file['currency'] = b_file['Currency']
        f_file['4.1.1'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.1'].values
        f_file['4.1.2'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.2'].values
        f_file['4.1.3'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.3'].values
        f_file['4.1.4'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.4'].values
        f_file['4.1.5'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.5'].values
        f_file['4.1.6'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.6'].values
        f_file['4.1.7'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.7'].values
        f_file['4.1.8'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.8'].values
        f_file['4.1.9'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.9'].values
        f_file['4.1.10'] = b_file[b_file.ReportLevel == 'Default Fund']['4.1.10'].values
    
    # 4.4.7.PDA & 4.4.7.MA
    try:
        f_file['4.4.7.PDA'] = a_file[a_file.Description == 'PeakDayAmountInPast12Months'].reset_index()['4_4_7']
        f_file['4.4.7.MA'] = a_file[a_file.Description == 'MeanAverageOverPrevious12Months'].reset_index()['4_4_7']
    
    #6.1.1.CLIENT & 6.1.1.TOTAL
        f_file['6.1.1.CLIENT'] = sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & c_file.ReportLevelIdentifier.isin(['Eurex Clearing'])]['6_1_1'])
        f_file['6.1.1.TOTAL'] = sum(c_file[c_file.Description.isin(['Total']) & c_file.ReportLevelIdentifier.isin(['Eurex Clearing'])]['6_1_1'])

    #6.2.15.PRE & POST
        f_file['6.2.15.PRE'] = d_file[d_file.Description == 'PreHaircut'].reset_index()['6_2_15']
        f_file['6.2.15.POST'] = d_file[d_file.Description == 'PostHaircut'].reset_index()['6_2_15']
    
    #number of CCP
        f_file['18.1.1.1'] = b_file['18_1_1_1']
        f_file['18.1.1.2'] = b_file['18_1_1_2']
    
    #number od CCP participants (average)
        f_file['18.1.2.2'] = b_file['18_1_2_2']
    
    #number of clients
        f_file['19.1.1'] = b_file['19_1_1']
        f_file['19.1.2'] = b_file['19_1_2']
      
    #for Q1-2016 & Q4-2015 there are differences
    except:
        f_file['4.4.7.PDA'] = a_file[a_file.Description == 'PeakDayAmountInPast12Months'].reset_index()['4.4.7']
        f_file['4.4.7.MA'] = a_file[a_file.Description == 'MeanAverageOverPrevious12Months'].reset_index()['4.4.7']
    
    #6.1.1.CLIENT & 6.1.1.TOTAL
        f_file['6.1.1.CLIENT'] = sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & c_file.ReportLevelIdentifier.isin(['Eurex Clearing'])]['6.1.1'])
        f_file['6.1.1.TOTAL'] = sum(c_file[c_file.Description.isin(['Total']) & c_file.ReportLevelIdentifier.isin(['Eurex Clearing'])]['6.1.1'])

    #6.2.15.PRE & POST
        f_file['6.2.15.PRE'] = d_file[d_file.Description == 'PreHaircut'].reset_index()['6.2.15']
        f_file['6.2.15.POST'] = d_file[d_file.Description == 'PostHaircut'].reset_index()['6.2.15']
    
    #number of CCP
        f_file['18.1.1.1'] = b_file['18.1.1.1']
        f_file['18.1.1.2'] = b_file['18.1.1.2']
    
    #number od CCP participants (average)
        f_file['18.1.2.2'] = b_file['18.1.2.2']
    
    #number of clients
        f_file['19.1.1'] = b_file['19.1.1']
        f_file['19.1.2'] = b_file['19.1.2']
    
    #23.1.2.OTC and 23.1.2.ETD and 23.2.1.OTC and 23.2.1.ETD
    f_file['23.1.2.OTC'] = None
    f_file['23.1.2.ETD'] = None
    f_file['23.2.1.OTC'] = None
    f_file['23.2.1.ETD'] = None
    
    ccp_save(ccp, quarter, f_file)
```
