```python
import numpy as np
import pandas as pd
import openpyxl 
import os

%matplotlib inline
import matplotlib.pyplot as plt  # Matlab-style plotting
import seaborn as sns
color = sns.color_palette()
sns.set_style('darkgrid')
import warnings
def ignore_warn(*args, **kwargs):
    pass
warnings.warn = ignore_warn #ignore annoying warning (from sklearn and seaborn)
```

```python
def ccp_oia_aggr_srt_count_1(quarter, col_name):
    ccp = pd.read_excel('../CCP/data/aio_aggr_sorted.xlsx', sheet_name='aio_aggr_sorted')
    ccp = ccp.set_index("CCP")
    ax = ccp[ccp['quarter'] == quarter][col_name].sort_values(ascending=False).plot.bar()
    ax.set_title(col_name, fontsize=20)
```

```python
ccp_oia_aggr_srt_count_1('Q2-2018', '18.1.1.1')
```

```python
def ccp_oia_aggr_srt_count_2(quarter, col_name):
    ccp = pd.read_excel('../CCP/data/aio_aggr_sorted.xlsx', sheet_name='aio_aggr_sorted')
    ax = sns.barplot(x="CCP", y=col_name, data=ccp[ccp['quarter'] == quarter].sort_values(by=col_name, ascending=False))
    ax.set_title(quarter, fontsize=12)
    plt.xticks(x="CCP", rotation='vertical')
```

```python
ccp_oia_aggr_srt_count_2('Q2-2018', '19.1.2')
```
