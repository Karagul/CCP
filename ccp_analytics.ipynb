```python
import numpy as np
import pandas as pd
import openpyxl 
import os

%matplotlib inline
import matplotlib.pyplot as plt  # Matlab-style plotting
import seaborn as sns
color = sns.color_palette()
sns.set_style('darkgrid')
import warnings
def ignore_warn(*args, **kwargs):
    pass
warnings.warn = ignore_warn #ignore annoying warning (from sklearn and seaborn)
```

```python
def ccp_count(quarter, col_name):
    ccp = pd.read_excel('../CCP/final_data/aio_aggr_sorted.xlsx', sheet_name='aio_aggr_sorted')
    ax = sns.barplot(x="CCP", y=col_name, data=ccp[ccp['quarter'] == quarter].sort_values(by=col_name, ascending=False))
    ax.set_title(quarter, fontsize=12)
    plt.xticks(x="CCP", rotation='vertical')
```

```python
def ccp_line(ccp_name, col_name):
    ccp = pd.read_excel('../CCP/final_data/aio_aggr_sorted.xlsx', sheet_name='aio_aggr_sorted')
    #ccp = ccp.set_index('quarter')
    ax = ccp[ccp['CCP'] == ccp_name].sort_values(by=['year', 'q']).reset_index()[col_name].plot.line()
    ax.set_title(ccp_name, fontsize=16)
    plt.ylabel(col_name, fontsize=16)
    #ax.set_ylim(ymin=0)
    ax.set_xticklabels([0, 'Q4-2015', 'Q2-2016', 'Q4-2016', 'Q2-2017', 'Q4-2017', 'Q2-2018'], rotation='vertical', fontsize=12)
```

```python
ccp_line('Eurex', '19.1.2')
```
