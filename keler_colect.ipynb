```python
import numpy as np
import pandas as pd
import openpyxl 
import os
%run ccp_func.ipynb
```

```python
def keler_colect(quarter):
    #quarter = format: 'Q3-2017'
    
    ccp = 'keler'
    q = quarter.split('-')
    num_q = q[0][1] + q[0][0]
    q = num_q + q[1]

    #for all keler files cut the last part of the name which is saving date.
    #basic
    b_file = pd.read_excel('../CCP/data/' + quarter.upper() + '/CPMI-IOSCO_Public-Disclosure_KELER-CCP_' 
                           + quarter.upper() + '.xlsx', sheet_name='AggregatedDataFile')
    
    #files in use
    a_file = pd.read_excel('../CCP/data/' + quarter.upper() + '/CPMI-IOSCO_Public-Disclosure_KELER-CCP_' 
                           + quarter.upper() + '.xlsx', 
                               sheet_name='KELER CCP_DataFile_4_4a')
    c_file = pd.read_excel('../CCP/data/' + quarter.upper() + '/CPMI-IOSCO_Public-Disclosure_KELER-CCP_' 
                           + quarter.upper() + '.xlsx', 
                               sheet_name='KELER CCP_DataFile_6_1')
    d_file = pd.read_excel('../CCP/data/' + quarter.upper() + '/CPMI-IOSCO_Public-Disclosure_KELER-CCP_' 
                           + quarter.upper() + '.xlsx', 
                               sheet_name='KELER CCP_DataFile_6.2')
    e_file = pd.read_excel('../CCP/data/' + quarter.upper() + '/CPMI-IOSCO_Public-Disclosure_KELER-CCP_' 
                           + quarter.upper() + '.xlsx', 
                               sheet_name='KELER CCP_DataFile_18')
    #some 23 are missing
    try:
        g_file = pd.read_excel('../CCP/data/' + quarter.upper() + '/CPMI-IOSCO_Public-Disclosure_KELER-CCP_' 
                           + quarter.upper() + '.xlsx', 
                               sheet_name='KELER CCP_DataFile_23_1')
    except:
        pass
    
    ex_rate = pd.read_excel('../CCP/data/kursy-walutowe-CCP.xlsx', sheet_name='Arkusz1')
    
    f_file = pd.DataFrame({'CCP':ccp.upper()}, index=[0,1,2,3])
    f_file['quarter'] = quarter
    f_file['clearning service'] = ['TEA', 'KGA', 'TP KGA', 'CEEGEX KGA']
    b_file = b_file.drop([0]).reset_index() #general description
    
    #str to int in 6.1.1
    for row in range(len(c_file)):
        if type(c_file.loc[row, '6.1.1']) == str:
            c_file.loc[row, '6.1.1'] = 0
        else:
            pass
    
    f_file['4.1.1'] = b_file['4.1.1']
    f_file['4.1.2'] = b_file['4.1.2']
    f_file['4.1.3'] = b_file['4.1.3']
    f_file['4.1.4'] = b_file['4.1.4']
    f_file['4.1.5'] = b_file['4.1.5']
    f_file['4.1.6'] = b_file['4.1.6']
    f_file['4.1.7'] = b_file['4.1.7']
    f_file['4.1.8'] = b_file['4.1.8']
    f_file['4.1.9'] = b_file['4.1.9']
    f_file['4.1.10'] = b_file['4.1.10']
    
    f_file['4.4.7.PDA'] = a_file[a_file.Description == 'PeakDayAmountInPast12Months'].reset_index()['4.4.7']
    f_file['4.4.7.MA'] = a_file[a_file.Description == 'MeanAverageOverPrevious12Months'].reset_index()['4.4.7']
    
    TEA_CLIENT= sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & (c_file.ReportLevelIdentifier == 'KELER CCP_TEA')]['6.1.1'])
    KGA_CLIENT= sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & (c_file.ReportLevelIdentifier == 'KELER CCP_KGA')]['6.1.1'])
    TP_CLIENT= sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & (c_file.ReportLevelIdentifier == 'KELER CCP_TP KGA')]['6.1.1'])
    C_CLIENT= sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & (c_file.ReportLevelIdentifier == 'KELER CCP_CEEGEX KGA')]['6.1.1'])

    f_file['6.1.1.CLIENT'] = [TEA_CLIENT, KGA_CLIENT, TP_CLIENT, C_CLIENT]
    
    f_file['6.1.1.TOTAL'] = c_file[c_file.Description == 'Total'].reset_index()['6.1.1']
    
    f_file['6.2.15.PRE'] = d_file[d_file.Description.isin(['TotalIM_PreHaircut'])]['6.2.15'].values
    f_file['6.2.15.POST'] = d_file[d_file.Description.isin(['TotalIM_PostHaircut'])]['6.2.15'].values
    
    f_file['18.1.1.1'] = b_file['18.1.1.1']
    f_file['18.1.1.2'] = b_file['18.1.1.2']

    f_file['18.1.2.2'] = b_file['18.1.2.2']

    f_file['19.1.1'] = b_file['19.1.1']
    f_file['19.1.2'] = b_file['19.1.2']
    
    try:
        #23.1.2.OTC & ETD
        TEA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP_TEA'])]['Average daily volume (# of trades)'])
        KGA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP_KGA'])]['Average daily volume (# of trades)'])
        TP_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP_TP KGA'])]['Average daily volume (# of trades)'])
        C_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP CEEGEX'])]['Average daily volume (# of trades)'])

        f_file['23.1.2.OTC'] = [TEA_OTC, KGA_OTC, TP_OTC, C_OTC]

        TEA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP_TEA'])]['Average daily volume (# of trades)'])
        KGA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP_KGA'])]['Average daily volume (# of trades)'])
        TP_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP_TP KGA'])]['Average daily volume (# of trades)'])
        C_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP CEEGEX'])]['Average daily volume (# of trades)'])

        f_file['23.1.2.ETD'] = [TEA_OTC, KGA_OTC, TP_OTC, C_OTC]


        #23.2.1.OTC & ETD
        TEA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP_TEA'])]['Average notional value of trades (EUR)'])
        KGA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP_KGA'])]['Average notional value of trades (EUR)'])
        TP_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP_TP KGA'])]['Average notional value of trades (EUR)'])
        C_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['OTD'])) & g_file['Clearing Service'].isin(['KELER CCP CEEGEX'])]['Average notional value of trades (EUR)'])

        f_file['23.2.1.OTC'] = [TEA_OTC, KGA_OTC, TP_OTC, C_OTC]

        TEA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP_TEA'])]['Average notional value of trades (EUR)'])
        KGA_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP_KGA'])]['Average notional value of trades (EUR)'])
        TP_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP_TP KGA'])]['Average notional value of trades (EUR)'])
        C_OTC = sum(g_file[(g_file['OTD/ETD/Other'].isin(['ETD'])) & g_file['Clearing Service'].isin(['KELER CCP CEEGEX'])]['Average notional value of trades (EUR)'])

        f_file['23.2.1.ETD'] = [TEA_OTC, KGA_OTC, TP_OTC, C_OTC]
    except:
        f_file['23.1.2.OTC'] = None
        f_file['23.1.2.ETD'] = None
        f_file['23.2.1.OTC'] = None
        f_file['23.2.1.ETD'] = None
    
    ccp_save(ccp, quarter, f_file)
```
