```python
import numpy as np
import pandas as pd
import openpyxl 
import os
%run ccp_func.ipynb
```

```python
def ice_nl_colect(quarter):
    #quarter = format: 'Q3-2017'
    
    ccp = 'ice clear netherlands'
    reverse = quarter.split('-')
    reverse = reverse[1] + reverse[0]
    
    #for all keler files cut the last part of the name which is saving date.
    #basic
    b_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICNL_AggregateDataFile_' 
                           + reverse.upper() + '.csv')
    #files in use 
    a_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICNL_DataFile_4_4a_' + reverse.upper() + '.csv')
    c_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICNL_DataFile_6_1_' + reverse.upper() + '.csv')
    d_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICNL_DataFile_6_2_' + reverse.upper() + '.csv')
    e_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICNL_DataFile_18_2_and_18_3_' + reverse.upper() + '.csv')
    g_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICNL_DataFile_23_' + reverse.upper() + '.csv')
    ex_rate = pd.read_excel('../CCP/data/kursy-walutowe-CCP.xlsx', sheet_name='Arkusz1')
    
    f_file = pd.DataFrame({'CCP':ccp.upper()}, index=[0])
    f_file['quarter'] = quarter
    f_file['clearning service'] = ['aggregated']
    
    try:
        f_file['4.1.1'] = b_file['4_1_1']
        f_file['4.1.2'] = b_file['4_1_2']
        f_file['4.1.3'] = b_file['4_1_3']
        f_file['4.1.4'] = b_file['4_1_4']
        f_file['4.1.5'] = b_file['4_1_5']
        f_file['4.1.6'] = b_file['4_1_6']
        f_file['4.1.7'] = b_file['4_1_7']
        f_file['4.1.8'] = b_file['4_1_8']
        f_file['4.1.9'] = b_file['4_1_9']
        f_file['4.1.10'] = b_file['4_1_10']
    except:
        #for q2-2016 and Q1-2016
        f_file['4.1.1'] = b_file['4.1.1']
        f_file['4.1.2'] = b_file['4.1.2']
        f_file['4.1.3'] = b_file['4.1.3']
        f_file['4.1.4'] = b_file['4.1.4']
        f_file['4.1.5'] = b_file['4.1.5']
        f_file['4.1.6'] = b_file['4.1.6']
        f_file['4.1.7'] = b_file['4.1.7']
        f_file['4.1.8'] = b_file['4.1.8']
        f_file['4.1.9'] = b_file['4.1.9']
        f_file['4.1.10'] = b_file['4.1.10']
    
    try:
        f_file['4.4.7.PDA'] = a_file[a_file.Description == 'PeakDayAmountInPast12Months'].reset_index()['4_4_7']
        f_file['4.4.7.MA'] = a_file[a_file.Description == 'MeanAverageOverPrevious12Months'].reset_index()['4_4_7']

        f_file['6.1.1.CLIENT'] = sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross'])]['6_1_1'])
        f_file['6.1.1.TOTAL'] = c_file[c_file.Description == 'Total'].reset_index()['6_1_1']

        f_file['6.2.15.PRE'] = d_file[d_file.Description.isin(['TotalIM_PreHaircut'])]['6_2_15'].values
        f_file['6.2.15.POST'] = d_file[d_file.Description.isin(['TotalIM_PostHaircut'])]['6_2_15'].values

        f_file['18.1.1.1'] = b_file['18_1_1_1']
        f_file['18.1.1.2'] = b_file['18_1_1_2']

        f_file['18.1.2.2'] = b_file['18_1_2_2']

        f_file['19.1.1'] = b_file['19_1_1']
        f_file['19.1.2'] = b_file['19_1_2']

        try:
            f_file['23.1.2.OTC'] = sum(g_file[g_file.Description == 'OTC']['23_1_2'])
            f_file['23.1.2.ETD'] = sum(g_file[g_file.Description == 'ETD']['23_1_2'])
        except:
            f_file['23.1.2.OTC'] = None
            f_file['23.1.2.ETD'] = None

        try:
            f_file['23.2.1.OTC'] = sum(g_file[g_file.Description == 'OTC']['23_2_1'])
            f_file['23.2.1.ETD'] = sum(g_file[g_file.Description == 'ETD']['23_2_1'])
        except:
            f_file['23.2.1.OTC'] = None
            f_file['23.2.1.ETD'] = None
    except:
        #for Q1-2016 and Q2-2016
        f_file['4.4.7.PDA'] = a_file[a_file.Description == 'PeakDayAmountInPast12Months'].reset_index()['4.4.7']
        f_file['4.4.7.MA'] = a_file[a_file.Description == 'MeanAverageOverPrevious12Months'].reset_index()['4.4.7']

        f_file['6.1.1.CLIENT'] = sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross'])]['6.1.1'])
        f_file['6.1.1.TOTAL'] = c_file[c_file.Description == 'Total'].reset_index()['6.1.1']

        f_file['6.2.15.PRE'] = d_file[d_file.Description.isin(['TotalIM_PreHaircut'])]['6.2.15'].values
        f_file['6.2.15.POST'] = d_file[d_file.Description.isin(['TotalIM_PostHaircut'])]['6.2.15'].values

        f_file['18.1.1.1'] = b_file['18.1.1.1']
        f_file['18.1.1.2'] = b_file['18.1.1.2']

        f_file['18.1.2.2'] = b_file['18.1.2.2']

        f_file['19.1.1'] = b_file['19.1.1']
        f_file['19.1.2'] = b_file['19.1.2']

        try:
            f_file['23.1.2.OTC'] = sum(g_file[g_file.Description == 'OTC']['23.1.2'])
            f_file['23.1.2.ETD'] = sum(g_file[g_file.Description == 'ETD']['23.1.2'])
        except:
            f_file['23.1.2.OTC'] = None
            f_file['23.1.2.ETD'] = None

        try:
            f_file['23.2.1.OTC'] = sum(g_file[g_file.Description == 'OTC']['23.2.1'])
            f_file['23.2.1.ETD'] = sum(g_file[g_file.Description == 'ETD']['23.2.1'])
        except:
            f_file['23.2.1.OTC'] = None
            f_file['23.2.1.ETD'] = None
    
    ccp_save(ccp, quarter, f_file)

```
