```python
def ice_europe_colect(quarter):
    #quarter = format: 'Q3-2017'
    
    ccp = 'ice clear europe'
    reverse = quarter.split('-')
    reverse = reverse[1] + reverse[0]
    
    #for all keler files cut the last part of the name which is saving date.
    try:
        #basic
        b_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_AggregatedDataFile_' 
                               + reverse.upper() + '.csv')
        #files in use 
        a_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_DataFile_4_4a_' + reverse.upper() + '.csv')
        c_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_DataFile_6_1_' + reverse.upper() + '.csv')
        d_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_DataFile_6_2_' + reverse.upper() + '.csv')
        e_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_DataFile_18_2_and_18_3_' + reverse.upper() + '.csv')
        g_file = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_DataFile_23_' + reverse.upper() + '.csv')
        
    except:
    #for Q3-2016 and lower
        b_filecds = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_CDS_AggregateDataFile_' 
                               + reverse.upper() + '.csv')
        b_filefo = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_F&O_AggregatedDataFile_' 
                               + reverse.upper() + '.csv')
        #files in use CDS
        a_filecds = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_CDS_DataFile_4_4a_' + reverse.upper() + '.csv')
        c_filecds = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_CDS_DataFile_6_1_' + reverse.upper() + '.csv')
        d_filecds = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_CDS_DataFile_6_2_' + reverse.upper() + '.csv')
        e_filecds = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_CDS_DataFile_18_2_and_18_3_' + reverse.upper() + '.csv')
        g_filecds = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_CDS_DataFile_23_' + reverse.upper() + '.csv')
        #files in use FO
        a_filefo = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_F&O_DataFile_4_4a_' + reverse.upper() + '.csv')
        c_filefo = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_F&O_DataFile_6_1_' + reverse.upper() + '.csv')
        d_filefo = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_F&O_DataFile_6_2_' + reverse.upper() + '.csv')
        e_filefo = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_F&O_DataFile_18_2_and_18_3_' + reverse.upper() + '.csv')
        g_filefo = pd.read_csv('../CCP/data/' + quarter.upper() + '/ICEU_F&O_DataFile_23_' + reverse.upper() + '.csv')
        
        
    ex_rate = pd.read_excel('../CCP/data/kursy-walutowe-CCP.xlsx', sheet_name='Arkusz1')
    
    try:  
        
        f_file = pd.DataFrame({'CCP':ccp.upper()}, index=[0, 1])
        f_file['quarter'] = quarter
        f_file['clearning service'] = ['F&O', 'CDS']
        
        #extra col 'ex_rate' in files
        add_ex_rate(b_file, ex_rate, quarter)

        #recalculate to EUR
        f_file['4.1.1'] = b_file['4_1_1'] // b_file['ex_rate']
        f_file['4.1.2'] = b_file['4_1_2'] // b_file['ex_rate']
        f_file['4.1.3'] = b_file['4_1_3'] // b_file['ex_rate']
        f_file['4.1.4'] = b_file['4_1_4'] // b_file['ex_rate'] 
        f_file['4.1.5'] = b_file['4_1_5'] // b_file['ex_rate'] 
        f_file['4.1.6'] = b_file['4_1_6'] // b_file['ex_rate']
        f_file['4.1.7'] = b_file['4_1_7'] // b_file['ex_rate']
        f_file['4.1.8'] = b_file['4_1_8'] // b_file['ex_rate']
        f_file['4.1.9'] = pd.to_numeric(b_file['4_1_9'], errors='coerce') // b_file['ex_rate']
        f_file['4.1.10'] = b_file['4_1_10'] // b_file['ex_rate']

        #extra col 'ex_rate' in files
        add_ex_rate(a_file, ex_rate, quarter)
        #recalculate to EUR
        a_file['4_4_7'] = a_file['4_4_7'] // a_file['ex_rate']

        f_file['4.4.7.PDA'] = a_file[(a_file.Description == 'PeakDayAmountInPast12Months')].reset_index()['4_4_7']
        f_file['4.4.7.MA'] = a_file[(a_file.Description == 'MeanAverageOverPrevious12Months')].reset_index()['4_4_7']

        #extra col 'ex_rate' in files
        add_ex_rate(c_file, ex_rate, quarter)
        #recalculate to EUR
        c_file['6_1_1'] = c_file['6_1_1'] // c_file['ex_rate']

        FQ = sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & (c_file.ReportLevelIdentifier == 'ICEU_F&O')]['6_1_1'])
        CDS = sum(c_file[c_file.Description.isin(['Client_Net', 'Client_Gross']) & (c_file.ReportLevelIdentifier == 'ICEU_CDS')]['6_1_1'])

        f_file['6.1.1.CLIENT'] = [FQ, CDS]
        f_file['6.1.1.TOTAL'] = c_file[c_file.Description == 'Total'].reset_index()['6_1_1']

        #extra col 'ex_rate' in files
        add_ex_rate(d_file, ex_rate, quarter)
        #recalculate to EUR
        d_file['6_2_15'] = d_file['6_2_15'] // d_file['ex_rate']

        f_file['6.2.15.PRE'] = d_file[d_file.Description.isin(['TotalIM_PreHaircut'])].reset_index()['6_2_15']
        f_file['6.2.15.POST'] = d_file[d_file.Description.isin(['TotalIM_PostHaircut'])].reset_index()['6_2_15']

        f_file['18.1.1.1'] = b_file['18_1_1_1']
        f_file['18.1.1.2'] = b_file['18_1_1_2']

        f_file['18.1.2.2'] = b_file['18_1_2_2']

        f_file['19.1.1'] = b_file['19_1_1']
        f_file['19.1.2'] = b_file['19_1_2']

        #extra col 'ex_rate' in files
        add_ex_rate(g_file, ex_rate, quarter)
        #recalculate to EUR
        g_file['23_1_2'] = g_file['23_1_2'] // g_file['ex_rate']
        g_file['23_2_1'] = pd.to_numeric(g_file['23_2_1'], errors='coerce') // g_file['ex_rate']

        FO_OTC = sum(g_file[(g_file.Description == 'OTC') & (g_file.ReportLevelIdentifier == 'ICEU_F&O')]['23_1_2'])
        CDS_OTC = sum(g_file[(g_file.Description == 'OTC') & (g_file.ReportLevelIdentifier == 'ICEU_CDS')]['23_1_2'])
        f_file['23.1.2.OTC'] = [FO_OTC, CDS_OTC]

        FO_ETD = sum(g_file[(g_file.Description == 'ETD') & (g_file.ReportLevelIdentifier == 'ICEU_F&O')]['23_1_2'])
        CDS_ETD = sum(g_file[(g_file.Description == 'ETD') & (g_file.ReportLevelIdentifier == 'ICEU_CDS')]['23_1_2'])
        f_file['23.1.2.ETD'] = [FO_ETD, CDS_ETD]

        #23.2.1
        FO_OTC = sum(g_file[(g_file.Description == 'OTC') & (g_file.ReportLevelIdentifier == 'ICEU_F&O')]['23_2_1'])
        CDS_OTC = sum(g_file[(g_file.Description == 'OTC') & (g_file.ReportLevelIdentifier == 'ICEU_CDS')]['23_2_1'])
        f_file['23.2.1.OTC'] = [FO_OTC, CDS_OTC]

        FO_ETD = sum(g_file[(g_file.Description == 'ETD') & (g_file.ReportLevelIdentifier == 'ICEU_F&O')]['23_2_1'])
        CDS_ETD = sum(g_file[(g_file.Description == 'ETD') & (g_file.ReportLevelIdentifier == 'ICEU_CDS')]['23_2_1'])
        f_file['23.2.1.ETD'] = [FO_ETD, CDS_ETD]
    except:
    #for Q3-2016 and below
        #CDS##################################################################################3
        f_filecds = pd.DataFrame({'CCP':ccp.upper()}, index=[0])
        f_filecds['quarter'] = quarter
        f_filecds['clearning service'] = ['CDS']
        #extra col 'ex_rate' in files
        add_ex_rate(b_filecds, ex_rate, quarter)

        #recalculate to EUR
        f_filecds['4.1.1'] = b_filecds['4_1_1'] // b_filecds['ex_rate']
        f_filecds['4.1.2'] = b_filecds['4_1_2'] // b_filecds['ex_rate']
        f_filecds['4.1.3'] = b_filecds['4_1_3'] // b_filecds['ex_rate']
        f_filecds['4.1.4'] = b_filecds['4_1_4'] // b_filecds['ex_rate'] 
        f_filecds['4.1.5'] = b_filecds['4_1_5'] // b_filecds['ex_rate'] 
        f_filecds['4.1.6'] = b_filecds['4_1_6'] // b_filecds['ex_rate']
        f_filecds['4.1.7'] = b_filecds['4_1_7'] // b_filecds['ex_rate']
        f_filecds['4.1.8'] = b_filecds['4_1_8'] // b_filecds['ex_rate']
        f_filecds['4.1.9'] = pd.to_numeric(b_filecds['4_1_9'], errors='coerce') // b_filecds['ex_rate']
        f_filecds['4.1.10'] = b_filecds['4_1_10'] // b_filecds['ex_rate']

        #extra col 'ex_rate' in files
        add_ex_rate(a_filecds, ex_rate, quarter)
        #recalculate to EUR
        a_filecds['4_4_7'] = a_filecds['4_4_7'] // a_filecds['ex_rate']

        f_filecds['4.4.7.PDA'] = a_filecds[(a_filecds.Description == 'PeakDayAmountInPast12Months')].reset_index()['4_4_7']
        f_filecds['4.4.7.MA'] = a_filecds[(a_filecds.Description == 'MeanAverageOverPrevious12Months')].reset_index()['4_4_7']

        #extra col 'ex_rate' in files
        add_ex_rate(c_filecds, ex_rate, quarter)
        #recalculate to EUR
        c_filecds['6_1_1'] = c_filecds['6_1_1'] // c_filecds['ex_rate']

        f_filecds['6.1.1.CLIENT'] = sum(c_filecds[c_filecds.Description.isin(['Client_Net', 'Client_Gross']) & (c_filecds.ReportLevelIdentifier == 'ICEU_CDS')]['6_1_1'])
        f_filecds['6.1.1.TOTAL'] = c_filecds[c_filecds.Description == 'Total'].reset_index()['6_1_1']

        #extra col 'ex_rate' in files
        add_ex_rate(d_filecds, ex_rate, quarter)
        #recalculate to EUR
        d_filecds['6_2_15'] = d_filecds['6_2_15'] // d_filecds['ex_rate']

        f_filecds['6.2.15.PRE'] = d_filecds[d_filecds.Description.isin(['TotalIM_PreHaircut'])].reset_index()['6_2_15']
        f_filecds['6.2.15.POST'] = d_filecds[d_filecds.Description.isin(['TotalIM_PostHaircut'])].reset_index()['6_2_15']

        f_filecds['18.1.1.1'] = b_filecds['18_1_1_1']
        f_filecds['18.1.1.2'] = b_filecds['18_1_1_2']

        f_filecds['18.1.2.2'] = b_filecds['18_1_2_2']

        f_filecds['19.1.1'] = b_filecds['19_1_1']
        f_filecds['19.1.2'] = b_filecds['19_1_2']

        #extra col 'ex_rate' in files
        add_ex_rate(g_filecds, ex_rate, quarter)
        #recalculate to EUR
        g_filecds['23_1_2'] = g_filecds['23_1_2'] // g_filecds['ex_rate']
        g_filecds['23_2_1'] = pd.to_numeric(g_filecds['23_2_1'], errors='coerce') // g_filecds['ex_rate']

        CDS_OTC = sum(g_filecds[(g_filecds.Description == 'OTC') & (g_filecds.ReportLevelIdentifier == 'ICEU_CDS')]['23_1_2'])
        f_filecds['23.1.2.OTC'] = [CDS_OTC]

        CDS_ETD = sum(g_filecds[(g_filecds.Description == 'ETD') & (g_filecds.ReportLevelIdentifier == 'ICEU_CDS')]['23_1_2'])
        f_filecds['23.1.2.ETD'] = [CDS_ETD]

        #23.2.1
        CDS_OTC = sum(g_filecds[(g_filecds.Description == 'OTC') & (g_filecds.ReportLevelIdentifier == 'ICEU_CDS')]['23_2_1'])
        f_filecds['23.2.1.OTC'] = [CDS_OTC]

        CDS_ETD = sum(g_filecds[(g_filecds.Description == 'OTD') & (g_filecds.ReportLevelIdentifier == 'ICEU_CDS')]['23_2_1'])
        f_filecds['23.2.1.ETD'] = [CDS_ETD]
        
        #FO#####################################################################################
        f_filefo = pd.DataFrame({'CCP':ccp.upper()}, index=[0])
        f_filefo['quarter'] = quarter
        f_filefo['clearning service'] = ['F&O']
        #extra col 'ex_rate' in files
        add_ex_rate(b_filefo, ex_rate, quarter)

        #recalculate to EUR
        f_filefo['4.1.1'] = b_filefo['4_1_1'] // b_filefo['ex_rate']
        f_filefo['4.1.2'] = b_filefo['4_1_2'] // b_filefo['ex_rate']
        f_filefo['4.1.3'] = b_filefo['4_1_3'] // b_filefo['ex_rate']
        f_filefo['4.1.4'] = b_filefo['4_1_4'] // b_filefo['ex_rate'] 
        f_filefo['4.1.5'] = b_filefo['4_1_5'] // b_filefo['ex_rate'] 
        f_filefo['4.1.6'] = b_filefo['4_1_6'] // b_filefo['ex_rate']
        f_filefo['4.1.7'] = b_filefo['4_1_7'] // b_filefo['ex_rate']
        f_filefo['4.1.8'] = b_filefo['4_1_8'] // b_filefo['ex_rate']
        f_filefo['4.1.9'] = pd.to_numeric(b_filefo['4_1_9'], errors='coerce') // b_filefo['ex_rate']
        f_filefo['4.1.10'] = b_filefo['4_1_10'] // b_filefo['ex_rate']

        #extra col 'ex_rate' in files
        add_ex_rate(a_filefo, ex_rate, quarter)
        #recalculate to EUR
        a_filefo['4_4_7'] = a_filefo['4_4_7'] // a_filefo['ex_rate']

        f_filefo['4.4.7.PDA'] = a_filefo[(a_filefo.Description == 'PeakDayAmountInPast12Months')].reset_index()['4_4_7']
        f_filefo['4.4.7.MA'] = a_filefo[(a_filefo.Description == 'MeanAverageOverPrevious12Months')].reset_index()['4_4_7']

        #extra col 'ex_rate' in files
        add_ex_rate(c_filefo, ex_rate, quarter)
        #recalculate to EUR
        c_filefo['6_1_1'] = c_filefo['6_1_1'] // c_filefo['ex_rate']

        f_filefo['6.1.1.CLIENT'] = sum(c_filefo[c_filecds.Description.isin(['Client_Net', 'Client_Gross']) & (c_filefo.ReportLevelIdentifier == 'ICEU_CDS')]['6_1_1'])
        f_filefo['6.1.1.TOTAL'] = c_filefo[c_filecds.Description == 'Total'].reset_index()['6_1_1']

        #extra col 'ex_rate' in files
        add_ex_rate(d_filefo, ex_rate, quarter)
        #recalculate to EUR
        d_filefo['6_2_15'] = d_filefo['6_2_15'] // d_filefo['ex_rate']

        f_filefo['6.2.15.PRE'] = d_filefo[d_filefo.Description.isin(['TotalIM_PreHaircut'])].reset_index()['6_2_15']
        f_filefo['6.2.15.POST'] = d_filefo[d_filefo.Description.isin(['TotalIM_PostHaircut'])].reset_index()['6_2_15']

        f_filefo['18.1.1.1'] = b_filefo['18_1_1_1']
        f_filefo['18.1.1.2'] = b_filefo['18_1_1_2']

        f_filefo['18.1.2.2'] = b_filefo['18_1_2_2']

        f_filefo['19.1.1'] = b_filefo['19_1_1']
        f_filefo['19.1.2'] = b_filefo['19_1_2']

        #extra col 'ex_rate' in files
        add_ex_rate(g_filefo, ex_rate, quarter)
        #recalculate to EUR
        g_filefo['23_1_2'] = g_filefo['23_1_2'] // g_filefo['ex_rate']
        g_filefo['23_2_1'] = pd.to_numeric(g_filefo['23_2_1'], errors='coerce') // g_filefo['ex_rate']

        CDS_OTC = sum(g_filefo[(g_filefo.Description == 'OTC') & (g_filefo.ReportLevelIdentifier == 'ICEU_F&O')]['23_1_2'])
        f_filefo['23.1.2.OTC'] = [CDS_OTC]

        CDS_ETD = sum(g_filefo[(g_filefo.Description == 'ETD') & (g_filefo.ReportLevelIdentifier == 'ICEU_F&O')]['23_1_2'])
        f_filefo['23.1.2.ETD'] = [CDS_ETD]

        #23.2.1
        CDS_OTC = sum(g_filefo[(g_filefo.Description == 'OTC') & (g_filefo.ReportLevelIdentifier == 'ICEU_F&O')]['23_2_1'])
        f_filefo['23.2.1.OTC'] = [CDS_OTC]

        CDS_ETD = sum(g_filefo[(g_filefo.Description == 'OTD') & (g_filefo.ReportLevelIdentifier == 'ICEU_F&O')]['23_2_1'])
        f_filefo['23.2.1.ETD'] = [CDS_ETD]
        
        f_file = pd.concat([f_filefo, f_filecds])
    
    ccp_save(ccp, quarter, f_file)
```
